```python
from pyspark import SparkContext
```

```python
sc = SparkContext()
```

```python
rdd = sc.textFile('/content/La+Celestina.txt')
```

```python
rdd.takeSample(False, 10)
```

El objetivo del proyecto es ver cuantas veces aparecen en el texto las siguientes palabras

- Celestina
- Calisto
- Melibea

```python
palabras = rdd.flatMap(lambda l: l.split(" ")).map(lambda s: s.strip(".,:;¿?¡!-\(\)\'\"")).countByValue()
```
